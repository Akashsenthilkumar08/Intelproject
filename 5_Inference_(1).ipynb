```python
from transformers import pipeline

# Replace this with your Hugging Face username/model name
model_id = "Sandeepmareeswaran/simple-qa-model"

# Load the fine-tuned Question Answering pipeline
qa_pipeline = pipeline("question-answering", model=model_id)

# Example context and question
context = """
Edsols Innovations is a company working in AI, IoT, and Robotics.
It provides EdTech solutions such as Reading Labs and training programs
for schools and colleges across India and the UAE.
"""

question = "What does Edsols Innovations work on?"

# Run inference
result = qa_pipeline(question=question, context=context)

print("Question:", question)
print("Answer:", result["answer"])
print("Score:", result["score"])
```
